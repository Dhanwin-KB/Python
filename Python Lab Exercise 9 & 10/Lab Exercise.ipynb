```python
# Line Graph

df = pd.read_csv('Social_Network_Ads.csv')
df.sort_values(by='Age', inplace=True)
age = df['Age']
salary = df['EstimatedSalary']
plt.figure(figsize=(10, 6))  
plt.plot(age, salary, linestyle='-', color='r') 
plt.title('Age vs Estimated Salary')  
plt.xlabel('Age')  
plt.ylabel('Estimated Salary')  
plt.grid(True)  
plt.show()
```

```python
# Heat Map

import seaborn as sns
df = pd.read_csv('Social_Network_Ads.csv')
filter = df.select_dtypes(include=['number']).drop(columns=['User ID'])
corr = filter.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Numeric Columns')
plt.show()
```

```python
#Scatter Plot

x = df['Age']
y = df['EstimatedSalary']
plt.figure(figsize=(8, 6))
plt.scatter(x, y, marker='o', color='b', alpha=0.5)
plt.title('Scatterplot of Age vs Estimated Salary')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.grid(True)
plt.show()
```
